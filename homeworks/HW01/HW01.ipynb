### Испорт библиотеки

```python
import numpy as np
```

### Создание массивов

```python
arr_list = np.array([10, 20, 30, 40, 50])

arr_zeros = np.zeros((3, 3))    # создание матрицы 3х3 со значениями 0

arr_range = np.arange(10)       # создания массива с цифрами от 0 до 9

print("Массив из списка:", arr_list)
print("\nМассив из нулей:\n", arr_zeros)
print("\nМассив из диапозона чисел: ", arr_range)
```

### Демонстрация атрибутов массива

```python
print("Форма массива (shape):", arr_zeros.shape)
print("Тип данных:          ", arr_zeros.dtype)
```

### Примеры индексирования и срезов

```python
element = arr_range[4]
sub_array = arr_range[3:6]  # получим эелменты с 3 по 6 не включая

print("Элмент с индексом 4:              ", element)
print("Срез массива (элементы с 3 по 6): ", sub_array)
```

### Примеры векторных операций и базовых статистик

```python
arr_mult = arr_list * 2         # умножаем каждый элемент массива на 2
arr_mean = np.mean(arr_list)    # считаем среднее значение

print("Массив умноженный на 2:", arr_mult)
print("Среднее значение:      ", arr_mean)
```
